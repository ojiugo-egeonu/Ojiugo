```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("C:/Users/Ojiugo/python_decal_sp26/course_assignments/homework7/star_data.txt", skiprows=1)

print(data.shape)
print(data[:5])
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]

print("Min magnitude:", np.min(magnitude))
print("Max magnitude:", np.max(magnitude))
print("Min temperature:", np.min(temperature))
print("Max temperature:", np.max(temperature))
```

```python
brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)

print("Brightest star index:", brightest)
print("Dimmest star index:", dimmest)
print("Hottest star index:", hottest)
print("Coldest star index:", coldest)
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]

print("Min magnitude:", np.min(magnitude))
print("Max magnitude:", np.max(magnitude))

print("Min temperature:", np.min(temperature))
print("Max temperature:", np.max(temperature))

brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)

print("Brightest star index:", brightest)
print("Dimmest star index:", dimmest)
print("Hottest star index:", hottest)
print("Coldest star index:", coldest)
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize=(8, 6))

plt.scatter(log_temperature, log_magnitude, label="Stars")

plt.scatter(log_temperature[brightest], log_magnitude[brightest],
            color="black", marker="*", s=300, label="Brightest Star")

plt.scatter(log_temperature[dimmest], log_magnitude[dimmest],
            color="purple", marker="*", s=300, label="Dimmest Star")

plt.scatter(log_temperature[hottest], log_magnitude[hottest],
            color="green", marker="*", s=300, label="Hottest Star")

plt.scatter(log_temperature[coldest], log_magnitude[coldest],
            color="orange", marker="*", s=300, label="Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.legend()

plt.show()
```
