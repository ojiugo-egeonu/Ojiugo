```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, 1)
    axes[0].plot(x, y, label=f"A={A}")

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y, label=f"w={w}")

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, label=f"A={A}, w={w}")

axes[0].set_title("Changing Amplitude", fontsize=14)
axes[0].set_xlabel("x", fontsize=12)
axes[0].set_ylabel("y", fontsize=12)
axes[0].legend()

axes[1].set_title("Changing Frequency", fontsize=14)
axes[1].set_xlabel("x", fontsize=12)
axes[1].set_ylabel("y", fontsize=12)
axes[1].legend()

axes[2].set_title("Changing Amplitude and Frequency", fontsize=14)
axes[2].set_xlabel("x", fontsize=12)
axes[2].set_ylabel("y", fontsize=12)
axes[2].legend()

plt.show()
```
